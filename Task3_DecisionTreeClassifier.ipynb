```python
columns = df[['age', 'campaign', 'duration']]

q1 = np.percentile(columns, 25, axis=0)
q3 = np.percentile(columns, 75, axis=0)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Mask out outliers
mask = (columns >= lower_bound) & (columns <= upper_bound)
df[['age', 'campaign', 'duration']] = columns.where(mask)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Evaluation function
def eval_model(y_test, y_pred):
    acc = accuracy_score(y_test, y_pred)
    print('Accuracy Score:', acc)
    
    cm = confusion_matrix(y_test, y_pred)
    print('Confusion Matrix:\n', cm)
    
    print('Classification Report:\n', classification_report(y_test, y_pred))

# Training/testing score function
def mscore(model):
    train_score = model.score(x_train, y_train)
    test_score = model.score(x_test, y_test)
    
    print('Training Score:', train_score)
    print('Testing Score:',test_score)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(criterion='gini', max_depth=5, min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
mscore(dt)
```

```python
ypred_dt=dt.predict(x_test)
print(ypred_dt)
```

```python
eval_model(y_test,ypred_dt)
```
